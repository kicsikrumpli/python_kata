```python
# 1. Iterative
from pprint import pprint

year_to_genres = dict()
for movie in movies:
  year = movie["year"]
  if year not in year_to_genres:
    year_to_genres[year] = dict()
  for genre in movie["genres"].split(","):
    if genre:
      year_to_genres[year][genre] = year_to_genres[year].get(genre, 0) + 1

# pprint(year_to_genres)

for year, genres in year_to_genres.items():
  most_popular_genre = max(genres.items(), key=lambda tup: tup[1])[0]
  print(f'{year}: {most_popular_genre}')
```

```python
# 2. Functional
from itertools import groupby
import json

# map movies dicts to year and genre list only, discard other attributes
year_genres = map(lambda m: {"year": m["year"], "genres": m["genres"].split(",")}, movies)

# comment out b/c otherwise the iterator is exhausted, 
# second run results in empty collection, 
# but we want to keep it lazy
# print(list(year_genres))

# sort by year. groupby expects sorted group key
year_genres_sorted = sorted(year_genres, key=lambda m: m["year"])
print('1. movies projected to year and genres, sorted by year:') 
pprint(year_genres_sorted) 
print('')

# group by year
by_year_genre_lists = {
    year: list(year_genres_list)
    for year, year_genres_list
    in groupby(year_genres_sorted, key=lambda movie: movie["year"])
}
print('2. movie genres grouped by year:')
pprint(by_year_genre_lists) 
print('')

# flatten genre lists
by_year_genre_list = {
    year: [
        genre.strip()
        for genres in map(lambda movie: movie["genres"], year_genres_list)
        for genre in genres
        if genre
    ]
    for year, year_genres_list
    in by_year_genre_lists.items()
}
print('3. all genre instances by year:')
pprint(by_year_genre_list)
print('')

# group by genres for each year, count distinct instances of each genre
by_year_genre_counts = {
    year: {
        genre: len(list(instances))
      for genre, instances
      in groupby(sorted(genre_list))
    }
    for year, genre_list
    in by_year_genre_list.items()
}

print('4. number of distinct instances of genres per year:') 
pprint(by_year_genre_counts)
print('')

print('-' * 10)
print('Most popular genres by year')
for year, genres in by_year_genre_counts.items():
  most_popular = max(genres, key=genres.get)
  print(f'{year}: {most_popular}')
```

```python
# 3. Pandas
import pandas as pd

# import into dataframe
movies_df = pd.DataFrame(data=movies)
print("\nmovies")
print(movies_df)

# split comma separated list in genres into list, turn each value into a separate row
genre_year = movies_df \
  .assign(genre=movies_df["genres"].str.split(',')) \
  .drop(["title", "rank", "roles", "directors", "genres"], axis=1) \
  .explode("genre")

# remove trailing and leading whitespace
genre_year["genre"] = genre_year["genre"].str.strip()

# remove empty values (where no genre was originally specified for the movie)
genre_year = genre_year[genre_year["genre"] != '']
print("\ngenres and years")
print(genre_year)

# count genres by year
genre_count_by_year = genre_year.groupby(["year", "genre"]).agg(count=("genre", "count"))
print("\ngenre counts by year")
print(genre_count_by_year)

# find the genre per year with max count
most_popular_genre_by_year = genre_count_by_year.reset_index().groupby(["year"]).agg({"genre": "max", "count": "max"})
print("\nmost popular genres by year:")
print(most_popular_genre_by_year)
```
